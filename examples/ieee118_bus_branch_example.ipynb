```python
import importlib
cim_profile = 'cimhub_2023'
cim = importlib.import_module('cimgraph.data_profile.' + cim_profile)
```

```python
from cimgraph.databases import Parameter, ConnectionParameters
from cimgraph.databases.blazegraph.blazegraph import BlazegraphConnection
# from cimgraph.databases.gridappsd import GridappsdConnection, get_topology_response
from cimgraph.models import BusBranchModel
import json
import time
```

```python
# Blazegraph connection for running outside the container
params = ConnectionParameters(url = "http://localhost:8889/bigdata/namespace/kb/sparql", iec61970_301=8, cim_profile=cim_profile)
bg = BlazegraphConnection(params)
```

```python
model_mrid = "1783D2A8-1204-4781-A0B4-7A73A2FA6038" #IEEE 118 Bus
```

## Create Distributed CIM Model:

```python
container = cim.ConnectivityNodeContainer(mRID=model_mrid)
```

```python
start = time.perf_counter()
network = BusBranchModel(connection=bg, container=container, distributed=False)
build_time = time.perf_counter() - start
print(build_time)
```

```python
network.get_all_edges(cim.ACLineSegment)
network.pprint(cim.ACLineSegment)
```
